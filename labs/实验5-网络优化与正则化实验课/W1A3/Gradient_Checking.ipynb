## 3 - 梯度检验的原理

反向传播计算梯度 $\frac{\partial J}{\partial \theta}$，其中 $\theta$ 表示模型的参数，
$J$ 由前向传播和损失函数共同计算得到。

由于前向传播相对容易实现且不易出错，我们可以利用计算 $J$ 的代码来验证
计算 $\frac{\partial J}{\partial \theta}$ 的代码是否正确。

回顾导数（梯度）的定义：

$$\frac{\partial J}{\partial \theta} = \lim_{\varepsilon \to 0} \frac{J(\theta + \varepsilon) - J(\theta - \varepsilon)}{2 \varepsilon} \tag{1}$$

当 $\varepsilon$ 足够小时，上式右侧是对梯度的很好近似。

**梯度检验的思路**：
- 用公式 (1) 计算"数值近似梯度" `gradapprox`
- 用反向传播计算实际梯度 `grad`
- 用以下公式计算两者的相对差异：

$$\text{difference} = \frac{\| \text{gradapprox} - \text{grad} \|_2}{\| \text{gradapprox} \|_2 + \| \text{grad} \|_2} \tag{2}$$

- 若 `difference` $< 2 \times 10^{-7}$，则反向传播很可能是正确的；
- 若 `difference` 较大，则反向传播很可能存在 bug。

## 5 - N 维梯度检验

下面描述了设备故障检测模型的前向和反向传播结构：
一个三层神经网络，结构为 LINEAR → ReLU → LINEAR → ReLU → LINEAR → Sigmoid。
输入为 4 维传感器特征向量，输出为故障概率。

下面提供了 `forward_propagation_n` 和 `backward_propagation_n` 的实现代码
（其中 `n` 表示适用于 N 维情况）。
注意：`backward_propagation_n` 中**故意引入了一个 bug**，你的任务是用梯度检验发现它。


````python
def forward_propagation_n(X, Y, parameters):
    """
    实现 N 维情况下的前向传播并计算代价。
    网络结构：LINEAR -> ReLU -> LINEAR -> ReLU -> LINEAR -> Sigmoid
    
    参数:
    X          -- 训练集，形状为 (输入维度, 样本数)
    Y          -- 标签，形状为 (1, 样本数)
    parameters -- 包含参数的字典：
                    W1 形状 (5, 4)，b1 形状 (5, 1)
                    W2 形状 (3, 5)，b2 形状 (3, 1)
                    W3 形状 (1, 3)，b3 形状 (1, 1)
    
    返回:
    cost  -- 代价函数值（逻辑斯谛代价）
    cache -- 中间值的元组 (Z1, A1, W1, b1, Z2, A2, W2, b2, Z3, A3, W3, b3)
    """
    
    m = X.shape[1]
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    W3 = parameters["W3"]
    b3 = parameters["b3"]

    # LINEAR -> RELU -> LINEAR -> RELU -> LINEAR -> SIGMOID
    Z1 = np.dot(W1, X) + b1
    A1 = relu(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = relu(Z2)
    Z3 = np.dot(W3, A2) + b3
    A3 = sigmoid(Z3)

    # 计算代价
    log_probs = np.multiply(-np.log(A3), Y) + np.multiply(-np.log(1 - A3), 1 - Y)
    cost = 1. / m * np.sum(log_probs)
    
    cache = (Z1, A1, W1, b1, Z2, A2, W2, b2, Z3, A3, W3, b3)
    
    return cost, cache
````
